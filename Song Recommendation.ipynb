```python
songs.rename(columns = {'Artist Genres':'Genres'}, inplace = True)
songs.rename(columns = {'Track Name':'Song'}, inplace = True)
songs.rename(columns = {'Artist Name(s)':'Artist_Name(s)'}, inplace = True)
songs.rename(columns = {'Album Name':'Album_Name'}, inplace = True)
songs.rename(columns = {'Track URI':'link'}, inplace = True)
songs.rename(columns = {'Album Image URL':'Image'}, inplace = True)
```

```python
songs.head()
```

```python
songs.iloc[0].Label
```

```python
songs['Genres'] = songs['Genres'].apply(lambda x:x.split(","))
```

```python
songs.head()
```

```python
songs['Label'] = songs['Label'].apply(lambda x:x.split(","))
```

```python
songs.head(2)
```

```python
songs['Album_Name'] = songs['Album_Name'].apply(lambda x:x.split(","))
```

```python
songs.head(2)
```

```python
songs['Artist_Name(s)'] = songs['Artist_Name(s)'].apply(lambda x:x.split(","))
```

```python
songs['Artist_Name(s)'] = songs['Artist_Name(s)'].apply(lambda x : [i.replace(" ","") for i in x])
songs['Album_Name'] = songs['Album_Name'].apply(lambda x : [i.replace(" ","") for i in x])
songs['Genres'] = songs['Genres'].apply(lambda x : [i.replace(" ","") for i in x])
songs['Label'] = songs['Label'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
songs['tags'] = songs['Artist_Name(s)'] + songs['Album_Name'] + songs['Genres'] + songs['Label']
```

```python
songs.head(1)
```

```python
new_df = songs[['Song','tags','link','Image']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[2])),reverse= True,key = lambda x : x[1])
```

```python
def recommend(song):
    song_index = new_df[new_df['Song'] == song].index[0]
    distances = similarity[song_index]
    song_list =  sorted(list(enumerate(distances)),reverse= True,key = lambda x : x[1])[1:21]
    for i in song_list:
        print(new_df.iloc[i[0]].Song)
```

```python
recommend('I Know You Want Me (Calle Ocho)')
```
